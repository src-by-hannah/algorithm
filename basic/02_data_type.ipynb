# **정수 자료형**

## **1. Char 형**

<br>

[크기] 1byte (8bits)

<br>

**1) unsigned char**

<center>| $2^7$ | $2^6$ | $2^5$ | $2^4$ | $2^3$ | $2^2$ | $2^1$ | $2^0$ |</center>

[범위] 0 ~ 255

<br>

**2) signed char**

<center>| $-2^7$ | $2^6$ | $2^5$ | $2^4$ | $2^3$ | $2^2$ | $2^1$ | $2^0$ |</center>
[범위] -128 ~ 127

**실수 자료형의 성질**

1. 실수의 저장/연산 과정에서 반드시 오차가 발생할 수 밖에 없다. 

2. double에 long long 범위의 정수를 함부로 담으면 안된다.

3. 실수를 비교할 때는 등호를 사용하면 안된다. (오차 때문에)
