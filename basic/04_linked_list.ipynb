# **정의와 성질**

* **정의**

: 원소들을 저장할 때 그 다음 원소가 있는 위치를 포함시키는 방식으로 저장하는 자료구조

: 우리는 첫 번째 원소의 주소만 기억하고 있는 것임

: 즉, 첫 번째 원소를 기준으로 연결 연결 연결해서 찾아나감

<BR>

* **성질**

1. k번째 원소를 확인/변경하기 위해 $O(k)$가 필요함

2. 임의의 위치에 원소를 추가/제거는 $O(1)$

3. 원소들이 메모리 상에 연속해있지 않아 Cache hit rate가 낮지만 할당이 다소 쉬움

<br>

* **종류**

1. 단일 연결 리스트 (Singly linked list)

: 각 원소가 자신의 다음 원소의 주소를 들고 있는 연결 리스트

<br>

2. 이중 연결 리스트

: 각 원소가 자신의 이전 원소와 다음 원소의 주소를 둘 다 들ㄷ고 있음

: 즉, 단일 연결 리스트와 달리 자신의 이전 원소를 알 수 있음

<br> 

3. 원형 연결 리스트

: 끝이 처음과 연결되어 있음

: 그 사이의 원소는 단일 혹은 이중 연결 리스트의 형태를 띌 수 있음

* **배열 vs. 연결리스트**

*★ 면접 때 자주 물어봄*

> 배열과 연결리스트는 선형 자료구조! <br>
&nbsp;&nbsp;&nbsp;&nbsp;→ cf. 비선형 자료구조 = 트리, 그래프, 해쉬

<img src="https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2Fdf2K0E%2FbtqCn1orhw4%2FINP9hzG71wG8a8fgW5HeNk%2Fimg.png" width = 70%>

▶ 배열처럼 그 뒤의 원소를 모두 옮기지 않고, 주소 참조만 변경해주면 되므로 $O(1)$의 시간만 소요됨
> 단, 추가하고 싶은 원소의 주소를 알고있을 때만 $O(1)$ <br>
> k번째 원소 뒤에 추가해야할 경우, 우선적으로 k번째 원소로 이동해야하므로 $O(n)$의 시간이 추가로 걸림

▶ 연결리스트 임의의 위치에 있는 원소를 확인/변경 시 원소의 위치만 알고 있을 경우 연결 연결 연결 해서 찾아야 하므로 $O(n)$의 시간이 걸림

> k번 째 원소를 확인/변경하려면 $O(k)$의 시간이 걸림<br>
이때, n개의 원소로 이루어진 연결 리스트라고 하면, 평균적으로 $O(n/2)$의 시간이 걸리며<br>
이에 따라 $O(n)$의 시간이 소요된다고 볼 수 있음

▶ 연결리스트는 다음 원소의 주소값을 저장하는 공간이 추가적으로 필요하므로  $O(n)$을 갖게 됨

<BR>

* **사용 예**

: 메모장

: 즉, 임의의 위치에 원소를 추가하거나 제거하는 연산을 할 때 연결 리스트가 유리

# **기능과 구현**

*★ 면접에서 손코딩 구현 나올 수도 있음*

```python
# RERERERE
# 배열 풀이 - 시간초과 발생

# BOJ 1406 에디터

import sys

def main() :
  # 초기 문자열 입력
  # s = input() // 시간초과 발생
  s = sys.stdin.readline().rstrip()

  txt = list(s)
  cursor = len(txt)

  loop = int(sys.stdin.readline()) # 명령어 수

  for i in range(loop) :
    do = sys.stdin.readline().rstrip() # 명령어 입력

    # L을 입력했을 경우
    if 'L' in do :
      cursor -= 1

    # D를 입력했을 경우
    elif 'D' in do :
      cursor += 1

    # B를 입력했을 경우
    elif 'B' in do and cursor != 0 :
      txt.pop(cursor-1)
      cursor -= 1

    elif 'P' in do :
      tmp = do.split()
      txt.insert(cursor, tmp[1])
      cursor += 1

    if cursor < 0 : 
        cursor = 0

    if cursor > len(txt) :
        cursor = len(txt)

    print(cursor)
    print(txt) 
  
  print(txt)

if __name__ == "__main__":
    main()
```
