시간복잡도 : $O(n)$

→ 한 번의 for 문

```python
def func1(n) :
  sum = 0 # 저장할 공간

  for i in range (1, n + 1, 1) : # "n 이하 / 자연수" 이므로 for문의 범위는 1 <= i <= n+1
    if i%3 == 0 or i%5 == 0 :
      sum += i

  return sum

print(func1(16))
print(func1(34567))
print(func1(27639))
```

시간복잡도 : $O(root(n))$

```python
def func3(n) :
  i = 1
  while (i*i <= n) :
    if (i*i == n) :
      return 1
      
    i += 1

  return 0

print(func3(9))
print(func3(693953651))
print(func3(756580036))
```
